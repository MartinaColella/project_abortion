```python
data['Value'] = data['Value'].astype(int)
```

```python
data
```

```python
data_pivot = data.pivot_table(values = ['Value'],
                             index = ['Age', 'marital_status', 'year'], 
                            columns = ['Work status']).reset_index()
```

```python
data_pivot
```

```python
data = data_pivot[data_pivot['marital_status'] != 'total']
```

```python
data
```

```python
data.columns = ['age', 'marital_status','year', 'employed', 'houswife', 'n_a', 'retired_or_other', 'student', 'total', 'unemployed_with_experience', 'unemployed_without_experience']
```

```python
data.drop(columns = 'total', inplace = True)
```

```python
data
```

```python
data_melt = data.melt(id_vars = ['age', 'marital_status', 'year'],
                    value_vars = ['employed', 'houswife', 'n_a', 'retired_or_other', 'student', 'unemployed_with_experience', 'unemployed_without_experience'], 
                     var_name = 'work_status', 
                     value_name = 'value')
```

```python
data_melt
```

```python
data_melt.to_csv('../clean_data/02a_employment_marital_long.csv', index = False)
```

id_vars: Column(s) to use as identifier variables
value_vars: Column(s) to unpivot. If not specified, uses all columns that are not set as id_vars.
var_name: Name to use for the ‘variable’ column.
value_name: Name to use for the ‘value’ column.


```python
data.to_csv('../clean_data/02_employment_marital_status.csv', index = False)
```
