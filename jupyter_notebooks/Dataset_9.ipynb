```python
import pandas as pd
```

```python
data = pd.read_csv('../raw/09_DCIS_IVG_CARATTEV1_09032023122858605.csv')
```

```python
italia = data[data['Territorio'] == 'Italia']
```

```python
italia = italia[italia['Tipo di intervento'] != 'totale']
```

```python
italia = italia[italia['Tipo dato'] == 'interruzioni volontarie della gravidanza']
```

```python
italia
```

```python
#italia['Value'] = italia['Value'].dropna
```

```python
italia['Value'] = italia['Value'].astype(int)
```

```python
italia = italia[italia['Settimane di amenorrea'] == 'totale']
```

```python
italia.drop(columns = ['ITTER107', 'TIPO_DATO14', 'TERANT', 'DURATA_A', 'DURATA_B', 'Flag Codes', 'Flags'], inplace = True)
```

```python
italia_pivoted = italia.pivot_table(values = ['Value'],
                            index = ['Territorio', 'TIME'], 
                            columns = ['Tipo di intervento']).reset_index()
```

```python
italia_pivoted
```

```python
italia_pivoted.columns
```

```python
italia_pivoted.columns = ['territory', 'year', 'other kind of vacuum aspiration', 'different from others', 'Karman method', 'not defined', 'dilatation and curettage', 'mifepristone and prostaglandina', 'prostaglandina only','mifepristone only']
```

```python
italia_pivoted
```

```python
italia_melted = italia_pivoted.melt(id_vars = ['territory', 'year'],
    value_vars = ['other kind of vacuum aspiration', 'different from others', 'Karman method', 'not defined', 'dilatation and curettage', 'mifepristone and prostaglandina', 'prostaglandina only','mifepristone only'],
          value_name = 'value')
```

```python
italia_melted.columns = ['territory', 'year', 'abortion_type', 'value']
```

```python
italia_melted
```

```python
italia_melted.to_csv('../clean_data/09_abortion_type_italy.csv', index = False)
```
