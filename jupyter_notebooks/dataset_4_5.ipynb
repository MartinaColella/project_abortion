```python
import pandas as pd
```

```python
data = pd.read_csv('../raw/07_DCIS_IVG_CARATTDON_05032023193841489.csv')
```

```python
data
```

```python
data = data[data['Data type'] == 'induced abortions']
```

```python
data
```

```python
# hot to change just some values of a column 
data.replace({'Place of residence' : {'Torino':'Piemonte',
                                      'Vercelli' : 'Piemonte', 
                                      'Biella' : 'Piemonte', 
                                      'Verbano-Cusio-Ossola': 'Piemonte', 
                                      'Novara' : 'Piemonte',
                                      'Cuneo': 'Piemonte', 
                                      'Asti': 'Piemonte', 
                                      'Alessandria' : 'Piemonte'}}, inplace = True)
```

```python
data['Place of residence'].unique()
```

```python
data[data['Place of residence'] == ''
```

```python
data_regions = data[data['Place of residence'].isin(['Piemomte', "Valle d'Aosta / Vallée d'Aoste", 'Liguria', 'Lombardia', 'Trentino Alto Adige / Südtirol', 'Veneto', 'Friuli-Venezia Giulia', 'Emilia-Romagna', 'Toscana', 'Umbria', 'Marche', 'Abruzzo', 'Molise', 'Campania', 'Puglia', 'Basilicata', 'Calabria', 'Sicilia', 'Sardegna', 'Abroad', 'da precisare'])]
```

```python
data_regions
```

```python
# pivot
data_pivot = data_regions.pivot_table(values = ['Value'],
                 index = ['Age and age class', 'TIME', ], 
                 columns = ['Place of residence']).reset_index()
```

```python
data_pivot.columns
```

```python
data_pivot.columns = ['age', 'year', 'abroad', 'abruzzo', 'basilicata', 'calabria', 'campania', 'Emilia_romagna', 'friuli_venezia_giulia', 'liguria', 'lombardia', 'marche', 'molise', 'puglia', 'sardegna', 'sicilia', 'toscana', 'trentino_alto_adige', 'umbria', "valle_d'aosta", 'veneto', 'undefined']
```

```python
tot = data_pivot[data_pivot['age'] == 'total']
```

```python
tot
```

```python
tot.to_csv('../clean_data/08_region_residence_tot_age.csv')
```

```python
data_pivot.to_csv('../clean_data/04_regions_age_with_tot.csv')
```

```python
data_no_tot = data_pivot[data_pivot['age'] != 'total']
data_no_tot
```

```python
data_no_tot.to_csv('../clean_data/09_region_residence_no_tot.csv')
```
